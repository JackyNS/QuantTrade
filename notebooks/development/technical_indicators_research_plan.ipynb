**ç›®æ ‡**: ä¸äº¤æ˜“ç­–ç•¥æ·±åº¦é›†æˆ

```python
# æŠ€æœ¯æŒ‡æ ‡å¼€å‘æ¡†æ¶

class IndicatorDevelopmentFramework:
    """æŠ€æœ¯æŒ‡æ ‡å¼€å‘å’Œæµ‹è¯•æ¡†æ¶"""
    
    def __init__(self):
        self.indicators = TechnicalIndicators()
        self.test_data = None
        
    def create_test_data(self, days=252):
        """ç”Ÿæˆæµ‹è¯•æ•°æ®"""
        np.random.seed(42)
        dates = pd.date_range('2024-01-01', periods=days, freq='D')
        
        # æ¨¡æ‹Ÿä»·æ ¼æ•°æ®
        price = 100
        prices = [price]
        
        for _ in range(days-1):
            change = np.random.normal(0, 0.02)  # 2%æ—¥æ³¢åŠ¨
            price = price * (1 + change)
            prices.append(price)
        
        # ç”ŸæˆOHLCVæ•°æ®
        high = [p * (1 + abs(np.random.normal(0, 0.01))) for p in prices]
        low = [p * (1 - abs(np.random.normal(0, 0.01))) for p in prices]
        volume = [int(np.random.uniform(1000000, 5000000)) for _ in prices]
        
        self.test_data = pd.DataFrame({
            'open': prices,
            'high': high,
            'low': low, 
            'close': prices,
            'volume': volume
        }, index=dates)
        
        return self.test_data
    
    def test_indicator(self, indicator_func, *args, **kwargs):
        """æµ‹è¯•å•ä¸ªæŒ‡æ ‡"""
        if self.test_data is None:
            self.create_test_data()
            
        try:
            result = indicator_func(*args, **kwargs)
            
            print(f"âœ… æŒ‡æ ‡æµ‹è¯•é€šè¿‡: {indicator_func.__name__}")
            print(f"   æ•°æ®å½¢çŠ¶: {result.shape if hasattr(result, 'shape') else 'scalar'}")
            print(f"   æ•°æ®ç±»å‹: {type(result)}")
            
            if hasattr(result, 'describe'):
                print(f"   åŸºæœ¬ç»Ÿè®¡:")
                print(result.describe())
            
            return True, result
            
        except Exception as e:
            print(f"âŒ æŒ‡æ ‡æµ‹è¯•å¤±è´¥: {indicator_func.__name__} - {e}")
            return False, None
    
    def benchmark_performance(self, indicator_func, iterations=100):
        """æ€§èƒ½åŸºå‡†æµ‹è¯•"""
        import time
        
        if self.test_data is None:
            self.create_test_data()
        
        # å‡†å¤‡å‚æ•°
        args = [self.test_data['close']] if 'close' in self.test_data.columns else []
        
        times = []
        for _ in range(iterations):
            start = time.time()
            try:
                indicator_func(*args)
                end = time.time()
                times.append(end - start)
            except:
                continue
        
        if times:
            avg_time = np.mean(times)
            print(f"âš¡ {indicator_func.__name__} æ€§èƒ½åŸºå‡†:")
            print(f"   å¹³å‡æ‰§è¡Œæ—¶é—´: {avg_time:.4f}ç§’")
            print(f"   æµ‹è¯•è¿­ä»£: {len(times)}æ¬¡")
            
            # æ€§èƒ½ç­‰çº§
            if avg_time < 0.001:
                level = "ğŸŸ¢ æå¿«"
            elif avg_time < 0.01:
                level = "ğŸŸ¡ å¿«"
            elif avg_time < 0.1:
                level = "ğŸŸ  ä¸­ç­‰"
            else:
                level = "ğŸ”´ æ…¢"
            
            print(f"   æ€§èƒ½ç­‰çº§: {level}")
            return avg_time
        else:
            print(f"âŒ æ— æ³•å®Œæˆæ€§èƒ½æµ‹è¯•: {indicator_func.__name__}")
            return None

# åˆå§‹åŒ–å¼€å‘æ¡†æ¶
framework = IndicatorDevelopmentFramework()
print("ğŸ”§ æŠ€æœ¯æŒ‡æ ‡å¼€å‘æ¡†æ¶å·²åˆå§‹åŒ–")
```
