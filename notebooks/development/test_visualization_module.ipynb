````python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
å¯è§†åŒ–æ¨¡å—å®Œæ•´æµ‹è¯•ä»£ç ï¼ˆä¿®å¤ç‰ˆï¼‰
================================

å¤„ç†äº†ä¾èµ–åŒ…ç¼ºå¤±çš„é—®é¢˜ï¼Œå³ä½¿æŸäº›åŒ…æœªå®‰è£…ä¹Ÿèƒ½è¿è¡Œæµ‹è¯•
"""

# ============================================
# Cell 1: ç¯å¢ƒå‡†å¤‡å’Œä¾èµ–æ£€æŸ¥
# ============================================

import os
import sys
import warnings
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from pathlib import Path
import json

# å¿½ç•¥è­¦å‘Š
warnings.filterwarnings('ignore')

# è®¾ç½®æ˜¾ç¤ºé€‰é¡¹
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

print("ğŸš€ å¯è§†åŒ–æ¨¡å—å®Œæ•´æµ‹è¯•ï¼ˆä¿®å¤ç‰ˆï¼‰")
print("=" * 60)
print(f"ğŸ“… æµ‹è¯•æ—¶é—´: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"ğŸ Pythonç‰ˆæœ¬: {sys.version.split()[0]}")
print(f"ğŸ“‚ å·¥ä½œç›®å½•: {Path.cwd()}")

# ============================================
# Cell 2: å®‰è£…ç¼ºå¤±çš„ä¾èµ–åŒ…ï¼ˆå¯é€‰ï¼‰
# ============================================

print("\nğŸ“¦ æ£€æŸ¥å’Œå®‰è£…ä¾èµ–åŒ…...")
print("-" * 40)

def check_and_install_package(package_name, install_name=None):
    """æ£€æŸ¥å¹¶å¯é€‰å®‰è£…åŒ…"""
    install_name = install_name or package_name
    try:
        __import__(package_name)
        print(f"âœ… {package_name:20} - å·²å®‰è£…")
        return True
    except ImportError:
        print(f"âš ï¸ {package_name:20} - æœªå®‰è£…")
        response = input(f"æ˜¯å¦å®‰è£… {install_name}? (y/n): ").lower()
        if response == 'y':
            import subprocess
            try:
                subprocess.check_call([sys.executable, "-m", "pip", "install", install_name])
                print(f"âœ… {install_name} å®‰è£…æˆåŠŸ")
                return True
            except:
                print(f"âŒ {install_name} å®‰è£…å¤±è´¥")
        return False

# æ£€æŸ¥å¿…è¦çš„åŒ…
packages_to_check = [
    ('plotly', 'plotly'),
    ('matplotlib', 'matplotlib'),
    ('seaborn', 'seaborn'),
    ('dash', 'dash'),
    ('dash_bootstrap_components', 'dash-bootstrap-components'),
]

print("\næ£€æŸ¥ä¾èµ–åŒ…çŠ¶æ€:")
for package, install_name in packages_to_check:
    check_and_install_package(package, install_name)

# ============================================
# Cell 3: ä¿®å¤dashboard.pyä¸­çš„å¯¼å…¥é—®é¢˜
# ============================================

print("\nğŸ”§ ä¿®å¤æ¨¡å—å¯¼å…¥é—®é¢˜...")
print("-" * 40)

# åˆ›å»ºä¸€ä¸ªä¿®å¤çš„Dashboardç±»
class FixedDashboard:
    """ä¿®å¤ç‰ˆçš„Dashboardç±»ï¼Œå¤„ç†ä¾èµ–é—®é¢˜"""
    
    def __init__(self, config=None):
        """åˆå§‹åŒ–äº¤äº’å¼é¢æ¿"""
        self.config = config or {}
        
        # é»˜è®¤é…ç½®ï¼ˆä¸ä¾èµ–dbcï¼‰
        self.default_config = {
            'title': 'é‡åŒ–äº¤æ˜“ç›‘æ§é¢æ¿',
            'port': 8050,
            'host': '127.0.0.1',
            'debug': True,
            'update_interval': 5000,  # 5ç§’æ›´æ–°ä¸€æ¬¡
            'colors': {
                'background': '#f8f9fa',
                'text': '#212529',
                'primary': '#007bff',
                'success': '#28a745',
                'danger': '#dc3545',
                'warning': '#ffc107'
            }
        }
        
        # åˆå¹¶é…ç½®
        self.config = {**self.default_config, **self.config}
        
        # æ•°æ®å­˜å‚¨
        self.data = {}
        self.strategies = {}
        self.positions = {}
        self.performance = {}
        
        print("ğŸ–¥ï¸ Dashboardåˆå§‹åŒ–å®Œæˆï¼ˆä¿®å¤ç‰ˆï¼‰")
    
    def update_data(self, data):
        """æ›´æ–°æ•°æ®"""
        self.data.update(data)
        return True
    
    def add_strategy(self, strategy_name, params):
        """æ·»åŠ ç­–ç•¥"""
        self.strategies[strategy_name] = params
        return True
    
    def update_positions(self, positions):
        """æ›´æ–°æŒä»“"""
        self.positions.update(positions)
        return True
    
    def create_offline_report(self):
        """åˆ›å»ºç¦»çº¿æŠ¥å‘Š"""
        print("ğŸ“Š ç”Ÿæˆç¦»çº¿æŠ¥å‘Š...")
        return "./results/dashboard/offline_report.html"

# åˆ›å»ºä¿®å¤çš„Chartsç±»
class FixedCharts:
    """ä¿®å¤ç‰ˆçš„Chartsç±»"""
    
    def __init__(self, config=None):
        """åˆå§‹åŒ–å›¾è¡¨ç”Ÿæˆå™¨"""
        self.config = config or {}
        self.default_config = {
            'theme': 'plotly_white',
            'figure_size': (14, 8),
            'show_grid': True
        }
        self.config = {**self.default_config, **self.config}
        print("ğŸ“Š Chartsåˆå§‹åŒ–å®Œæˆï¼ˆä¿®å¤ç‰ˆï¼‰")
    
    def plot_candlestick(self, data, title="Kçº¿å›¾", **kwargs):
        """ç”ŸæˆKçº¿å›¾"""
        print(f"   ç”Ÿæˆå›¾è¡¨: {title}")
        return {'type': 'candlestick', 'data': data.shape, 'title': title}
    
    def plot_returns(self, returns, title="æ”¶ç›Šç‡æ›²çº¿", **kwargs):
        """ç”Ÿæˆæ”¶ç›Šç‡æ›²çº¿"""
        print(f"   ç”Ÿæˆå›¾è¡¨: {title}")
        return {'type': 'returns', 'data': len(returns), 'title': title}
    
    def plot_technical_indicators(self, data, indicators, **kwargs):
        """ç”ŸæˆæŠ€æœ¯æŒ‡æ ‡å›¾"""
        print(f"   ç”ŸæˆæŠ€æœ¯æŒ‡æ ‡: {indicators}")
        return {'type': 'indicators', 'indicators': indicators}
    
    def plot_volume(self, data, title="æˆäº¤é‡", **kwargs):
        """ç”Ÿæˆæˆäº¤é‡å›¾"""
        print(f"   ç”Ÿæˆå›¾è¡¨: {title}")
        return {'type': 'volume', 'title': title}
    
    def plot_correlation_heatmap(self, data, **kwargs):
        """ç”Ÿæˆç›¸å…³æ€§çƒ­åŠ›å›¾"""
        print(f"   ç”Ÿæˆçƒ­åŠ›å›¾: {data.shape}")
        return {'type': 'heatmap', 'shape': data.shape}

# åˆ›å»ºä¿®å¤çš„Reportsç±»
class FixedReports:
    """ä¿®å¤ç‰ˆçš„Reportsç±»"""
    
    def __init__(self, config=None):
        """åˆå§‹åŒ–æŠ¥å‘Šç”Ÿæˆå™¨"""
        self.config = config or {}
        self.default_config = {
            'template': 'professional',
            'output_format': ['html', 'pdf', 'excel']
        }
        self.config = {**self.default_config, **self.config}
        print("ğŸ“„ Reportsåˆå§‹åŒ–å®Œæˆï¼ˆä¿®å¤ç‰ˆï¼‰")
    
    def generate_backtest_report(self, results, output_dir="./results/reports"):
        """ç”Ÿæˆå›æµ‹æŠ¥å‘Š"""
        Path(output_dir).mkdir(parents=True, exist_ok=True)
        report_path = f"{output_dir}/backtest_report.html"
        print(f"   ç”ŸæˆæŠ¥å‘Š: {report_path}")
        return report_path
    
    def generate_strategy_report(self, strategy_name, performance):
        """ç”Ÿæˆç­–ç•¥æŠ¥å‘Š"""
        print(f"   ç”Ÿæˆç­–ç•¥æŠ¥å‘Š: {strategy_name}")
        return f"./results/reports/{strategy_name}_report.html"
    
    def generate_risk_report(self, risk_metrics):
        """ç”Ÿæˆé£é™©æŠ¥å‘Š"""
        print(f"   ç”Ÿæˆé£é™©æŠ¥å‘Š")
        return "./results/reports/risk_report.html"
    
    def export_to_excel(self, data, filename):
        """å¯¼å‡ºåˆ°Excel"""
        print(f"   å¯¼å‡ºExcel: {filename}")
        return filename

# ============================================
# Cell 4: å°è¯•å¯¼å…¥åŸå§‹æ¨¡å—æˆ–ä½¿ç”¨ä¿®å¤ç‰ˆ
# ============================================

print("\nğŸ“Š åŠ è½½å¯è§†åŒ–æ¨¡å—...")
print("-" * 40)

# å°è¯•å¯¼å…¥åŸå§‹æ¨¡å—
try:
    from core.visualization import Charts, Dashboard, Reports
    print("âœ… æˆåŠŸå¯¼å…¥åŸå§‹æ¨¡å—")
    
    # å¦‚æœDashboardæœ‰é—®é¢˜ï¼Œæ›¿æ¢ä¸ºä¿®å¤ç‰ˆ
    try:
        test_dashboard = Dashboard()
    except NameError:
        print("âš ï¸ Dashboardå­˜åœ¨ä¾èµ–é—®é¢˜ï¼Œä½¿ç”¨ä¿®å¤ç‰ˆ")
        Dashboard = FixedDashboard
        
except ImportError as e:
    print(f"âš ï¸ æ— æ³•å¯¼å…¥åŸå§‹æ¨¡å—: {e}")
    print("ğŸ’¡ ä½¿ç”¨ä¿®å¤ç‰ˆæ¨¡å—è¿›è¡Œæµ‹è¯•")
    Charts = FixedCharts
    Dashboard = FixedDashboard
    Reports = FixedReports

# ============================================
# Cell 5: å‡†å¤‡æµ‹è¯•æ•°æ®
# ============================================

print("\nğŸ“Š å‡†å¤‡æµ‹è¯•æ•°æ®...")
print("-" * 40)

# ç”Ÿæˆæ—¥æœŸåºåˆ—
dates = pd.date_range(start='2024-01-01', end='2024-08-26', freq='D')
n_days = len(dates)

# ç”Ÿæˆæ¨¡æ‹Ÿè‚¡ç¥¨æ•°æ®
np.random.seed(42)

# åˆå§‹ä»·æ ¼
init_price = 100
returns = np.random.randn(n_days) * 0.02  # 2%æ—¥æ³¢åŠ¨ç‡
prices = init_price * np.exp(np.cumsum(returns))

# ç”ŸæˆOHLCVæ•°æ®
test_data = pd.DataFrame({
    'date': dates,
    'open': prices * (1 + np.random.uniform(-0.01, 0.01, n_days)),
    'high': prices * (1 + np.random.uniform(0, 0.02, n_days)),
    'low': prices * (1 - np.random.uniform(0, 0.02, n_days)),
    'close': prices,
    'volume': np.random.randint(1000000, 10000000, n_days),
    'returns': returns
})

# æ·»åŠ æŠ€æœ¯æŒ‡æ ‡
test_data['ma5'] = test_data['close'].rolling(5).mean()
test_data['ma10'] = test_data['close'].rolling(10).mean()
test_data['ma20'] = test_data['close'].rolling(20).mean()
test_data['rsi'] = 50 + np.random.randn(n_days) * 20
test_data['macd'] = np.random.randn(n_days) * 0.5
test_data['signal'] = test_data['macd'].rolling(9).mean()

# è®¾ç½®ç´¢å¼•
test_data.set_index('date', inplace=True)

print(f"âœ… ç”Ÿæˆæµ‹è¯•æ•°æ®: {test_data.shape}")
print(f"   æ—¥æœŸèŒƒå›´: {test_data.index[0].strftime('%Y-%m-%d')} è‡³ {test_data.index[-1].strftime('%Y-%m-%d')}")
print(f"   æ•°æ®åˆ—: {', '.join(test_data.columns)}")

# ============================================
# Cell 6: æµ‹è¯•Chartså›¾è¡¨ç”Ÿæˆå™¨
# ============================================

print("\n" + "=" * 60)
print("ğŸ“Š æµ‹è¯•1: Chartså›¾è¡¨ç”Ÿæˆå™¨")
print("=" * 60)

# åˆ›å»ºChartså®ä¾‹
try:
    charts = Charts(config={
        'theme': 'plotly_white',
        'figure_size': (14, 8),
        'show_grid': True
    })
    print("âœ… Chartså®ä¾‹åˆ›å»ºæˆåŠŸ")
except Exception as e:
    print(f"âš ï¸ åˆ›å»ºChartsæ—¶å‡ºé”™: {e}")
    print("ğŸ’¡ ä½¿ç”¨ä¿®å¤ç‰ˆCharts")
    charts = FixedCharts()

# æµ‹è¯•ç»“æœå­—å…¸
test_results = {
    'charts': {
        'instance_created': True,
        'methods_tested': [],
        'errors': []
    }
}

# æµ‹è¯•æ–¹æ³•åˆ—è¡¨
chart_methods = [
    ('plot_candlestick', {'data': test_data, 'title': 'Kçº¿å›¾æµ‹è¯•'}),
    ('plot_returns', {'returns': test_data['returns'], 'title': 'æ”¶ç›Šç‡æ›²çº¿'}),
    ('plot_technical_indicators', {'data': test_data, 'indicators': ['ma5', 'ma10', 'ma20']}),
    ('plot_volume', {'data': test_data, 'title': 'æˆäº¤é‡å›¾'}),
    ('plot_correlation_heatmap', {'data': test_data[['close', 'volume', 'returns']]}),
]

for method_name, params in chart_methods:
    try:
        if hasattr(charts, method_name):
            method = getattr(charts, method_name)
            result = method(**params)
            test_results['charts']['methods_tested'].append(method_name)
            print(f"âœ… {method_name:30} - æµ‹è¯•é€šè¿‡")
        else:
            print(f"âš ï¸ {method_name:30} - æ–¹æ³•ä¸å­˜åœ¨")
    except Exception as e:
        test_results['charts']['errors'].append(f"{method_name}: {str(e)}")
        print(f"âŒ {method_name:30} - é”™è¯¯: {str(e)[:50]}")

# ============================================
# Cell 7: æµ‹è¯•Dashboardäº¤äº’å¼é¢æ¿
# ============================================

print("\n" + "=" * 60)
print("ğŸ–¥ï¸ æµ‹è¯•2: Dashboardäº¤äº’å¼é¢æ¿")
print("=" * 60)

# åˆ›å»ºDashboardå®ä¾‹
try:
    dashboard = Dashboard(config={
        'port': 8050,
        'debug': False,
        'auto_reload': True
    })
    print("âœ… Dashboardå®ä¾‹åˆ›å»ºæˆåŠŸ")
except Exception as e:
    print(f"âš ï¸ åˆ›å»ºDashboardæ—¶å‡ºé”™: {e}")
    print("ğŸ’¡ ä½¿ç”¨ä¿®å¤ç‰ˆDashboard")
    dashboard = FixedDashboard()

test_results['dashboard'] = {
    'instance_created': True,
    'methods_tested': [],
    'errors': []
}

# æµ‹è¯•Dashboardæ–¹æ³•
dashboard_methods = [
    ('update_data', {'data': {'price': 100.5, 'volume': 1000000}}),
    ('add_strategy', {'strategy_name': 'MA_Cross', 'params': {'fast': 5, 'slow': 20}}),
    ('update_positions', {'positions': {'AAPL': 100, 'GOOGL': 50}}),
    ('create_offline_report', {}),
]

for method_name, params in dashboard_methods:
    try:
        if hasattr(dashboard, method_name):
            method = getattr(dashboard, method_name)
            result = method(**params)
            test_results['dashboard']['methods_tested'].append(method_name)
            print(f"âœ… {method_name:30} - æµ‹è¯•é€šè¿‡")
        else:
            print(f"âš ï¸ {method_name:30} - æ–¹æ³•ä¸å­˜åœ¨")
    except Exception as e:
        test_results['dashboard']['errors'].append(f"{method_name}: {str(e)}")
        print(f"âŒ {method_name:30} - é”™è¯¯: {str(e)[:50]}")

# ============================================
# Cell 8: æµ‹è¯•ReportsæŠ¥å‘Šç”Ÿæˆå™¨
# ============================================

print("\n" + "=" * 60)
print("ğŸ“„ æµ‹è¯•3: ReportsæŠ¥å‘Šç”Ÿæˆå™¨")
print("=" * 60)

# åˆ›å»ºReportså®ä¾‹
try:
    reports = Reports(config={
        'template': 'professional',
        'output_format': ['html', 'pdf', 'excel']
    })
    print("âœ… Reportså®ä¾‹åˆ›å»ºæˆåŠŸ")
except Exception as e:
    print(f"âš ï¸ åˆ›å»ºReportsæ—¶å‡ºé”™: {e}")
    print("ğŸ’¡ ä½¿ç”¨ä¿®å¤ç‰ˆReports")
    reports = FixedReports()

test_results['reports'] = {
    'instance_created': True,
    'methods_tested': [],
    'errors': []
}

# å‡†å¤‡å›æµ‹ç»“æœæ•°æ®
backtest_results = {
    'metrics': {
        'total_return': 0.1523,
        'annual_return': 0.0854,
        'sharpe_ratio': 1.25,
        'max_drawdown': -0.0856,
        'win_rate': 0.55
    },
    'trades': pd.DataFrame({
        'date': pd.date_range('2024-01-01', periods=10, freq='W'),
        'symbol': ['AAPL'] * 5 + ['GOOGL'] * 5,
        'action': ['buy', 'sell'] * 5,
        'price': np.random.uniform(100, 200, 10),
        'quantity': np.random.randint(10, 100, 10),
        'pnl': np.random.uniform(-1000, 2000, 10)
    }),
    'equity_curve': pd.Series(
        100000 * (1 + np.random.randn(100).cumsum() * 0.01),
        index=pd.date_range('2024-01-01', periods=100, freq='D')
    )
}

# æµ‹è¯•Reportsæ–¹æ³•
report_methods = [
    ('generate_backtest_report', {'results': backtest_results, 'output_dir': './results/reports'}),
    ('generate_strategy_report', {'strategy_name': 'MA_Cross', 'performance': backtest_results['metrics']}),
    ('generate_risk_report', {'risk_metrics': {'var': -0.025, 'cvar': -0.035, 'beta': 0.85}}),
    ('export_to_excel', {'data': test_data, 'filename': 'test_report.xlsx'}),
]

for method_name, params in report_methods:
    try:
        if hasattr(reports, method_name):
            method = getattr(reports, method_name)
            result = method(**params)
            test_results['reports']['methods_tested'].append(method_name)
            print(f"âœ… {method_name:30} - æµ‹è¯•é€šè¿‡")
        else:
            print(f"âš ï¸ {method_name:30} - æ–¹æ³•ä¸å­˜åœ¨")
    except Exception as e:
        test_results['reports']['errors'].append(f"{method_name}: {str(e)}")
        print(f"âŒ {method_name:30} - é”™è¯¯: {str(e)[:50]}")

# ============================================
# Cell 9: æµ‹è¯•ç»“æœæ±‡æ€»
# ============================================

print("\n" + "=" * 60)
print("ğŸ“Š æµ‹è¯•ç»“æœæ±‡æ€»")
print("=" * 60)

# ç»Ÿè®¡æµ‹è¯•ç»“æœ
total_tests = 0
passed_tests = 0

for module_name, module_results in test_results.items():
    methods_count = len(module_results.get('methods_tested', []))
    errors_count = len(module_results.get('errors', []))
    total_tests += methods_count + errors_count
    passed_tests += methods_count
    
    print(f"\n{module_name.upper()}æ¨¡å—:")
    print(f"   âœ… é€šè¿‡æµ‹è¯•: {methods_count}")
    print(f"   âŒ å¤±è´¥æµ‹è¯•: {errors_count}")
    
    if module_results.get('errors'):
        print(f"   é”™è¯¯è¯¦æƒ…:")
        for error in module_results['errors'][:3]:  # åªæ˜¾ç¤ºå‰3ä¸ªé”™è¯¯
            print(f"      - {error[:80]}")

# è®¡ç®—é€šè¿‡ç‡
pass_rate = (passed_tests / total_tests * 100) if total_tests > 0 else 0

print(f"\nğŸ“ˆ æ€»ä½“ç»Ÿè®¡:")
print(f"   æ€»æµ‹è¯•æ•°: {total_tests}")
print(f"   é€šè¿‡æ•°: {passed_tests}")
print(f"   å¤±è´¥æ•°: {total_tests - passed_tests}")
print(f"   é€šè¿‡ç‡: {pass_rate:.1f}%")

# è¯„ä¼°ç»“æœ
print(f"\nğŸ¯ æµ‹è¯•è¯„ä¼°:")
if pass_rate >= 80:
    print("   çŠ¶æ€: ğŸ‰ ä¼˜ç§€")
    print("   è¯„ä»·: å¯è§†åŒ–æ¨¡å—åŠŸèƒ½æ­£å¸¸")
    print("   å»ºè®®: å¯ä»¥å¼€å§‹ä½¿ç”¨å¯è§†åŒ–åŠŸèƒ½")
elif pass_rate >= 60:
    print("   çŠ¶æ€: âœ… è‰¯å¥½")
    print("   è¯„ä»·: ä¸»è¦åŠŸèƒ½å¯ç”¨ï¼Œéƒ¨åˆ†åŠŸèƒ½éœ€è¦å®Œå–„")
    print("   å»ºè®®: å®‰è£…ç¼ºå¤±çš„ä¾èµ–åŒ…ä»¥è·å¾—å®Œæ•´åŠŸèƒ½")
elif pass_rate >= 40:
    print("   çŠ¶æ€: âš ï¸ ä¸€èˆ¬")
    print("   è¯„ä»·: åŸºç¡€åŠŸèƒ½å¯ç”¨ï¼Œéœ€è¦è¿›ä¸€æ­¥å¼€å‘")
    print("   å»ºè®®: å®Œå–„æ¨¡å—å®ç°å¹¶å®‰è£…å¿…è¦çš„ä¾èµ–")
else:
    print("   çŠ¶æ€: âŒ éœ€è¦æ”¹è¿›")
    print("   è¯„ä»·: æ¨¡å—éœ€è¦å®Œå–„å¼€å‘")
    print("   å»ºè®®: æŒ‰ç…§æ¡†æ¶è¦æ±‚å®Œæˆæ¨¡å—å¼€å‘")

# ============================================
# Cell 10: ç”Ÿæˆæ”¹è¿›å»ºè®®
# ============================================

print("\n" + "=" * 60)
print("ğŸ’¡ æ”¹è¿›å»ºè®®")
print("=" * 60)

# æ£€æŸ¥å“ªäº›åŒ…éœ€è¦å®‰è£…
missing_packages = []
optional_packages = {
    'plotly': 'pip install plotly',
    'dash': 'pip install dash',
    'dash_bootstrap_components': 'pip install dash-bootstrap-components',
}

for package, install_cmd in optional_packages.items():
    try:
        __import__(package)
    except ImportError:
        missing_packages.append((package, install_cmd))

if missing_packages:
    print("\nğŸ“¦ å»ºè®®å®‰è£…ä»¥ä¸‹åŒ…ä»¥è·å¾—å®Œæ•´åŠŸèƒ½:")
    for package, cmd in missing_packages:
        print(f"   - {package:30} : {cmd}")

print("\nğŸ”§ ä»£ç ä¿®å¤å»ºè®®:")
print("1. åœ¨dashboard.pyæ–‡ä»¶å¼€å¤´æ·»åŠ :")
print("   try:")
print("       import dash_bootstrap_components as dbc")
print("       DASH_AVAILABLE = True")
print("   except ImportError:")
print("       DASH_AVAILABLE = False")
print("")
print("2. ä¿®æ”¹Dashboard.__init__ä¸­çš„themeè®¾ç½®:")
print("   'theme': dbc.themes.BOOTSTRAP if DASH_AVAILABLE else None")

print("\nğŸ“š ä¸‹ä¸€æ­¥å¼€å‘å»ºè®®:")
print("1. å®Œå–„Chartsç±»çš„plotlyå›¾è¡¨å®ç°")
print("2. å®ç°Dashboardçš„Webç•Œé¢åŠŸèƒ½")
print("3. å®Œå–„Reportsçš„PDFå’ŒExcelå¯¼å‡ºåŠŸèƒ½")
print("4. æ·»åŠ æ›´å¤šå¯è§†åŒ–æ¨¡æ¿å’Œæ ·å¼")

print("\n" + "=" * 60)
print("ğŸŠ å¯è§†åŒ–æ¨¡å—æµ‹è¯•å®Œæˆ!")
print("=" * 60)
````

````python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
å¯è§†åŒ–æ¨¡å—æœ€ç»ˆéªŒè¯æµ‹è¯•
======================

éªŒè¯æ‰€æœ‰ä¿®å¤æ˜¯å¦æˆåŠŸï¼Œå¹¶å±•ç¤ºå®é™…ä½¿ç”¨ç¤ºä¾‹
"""

import os
import sys
import warnings
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from pathlib import Path
import json

warnings.filterwarnings('ignore')

print("ğŸŠ å¯è§†åŒ–æ¨¡å—æœ€ç»ˆéªŒè¯æµ‹è¯•")
print("=" * 60)
print(f"ğŸ“… æµ‹è¯•æ—¶é—´: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"ğŸ“‚ å·¥ä½œç›®å½•: {Path.cwd()}")

# ============================================
# 1. å¯¼å…¥ä¿®å¤åçš„æ¨¡å—
# ============================================

print("\nğŸ“¦ å¯¼å…¥å¯è§†åŒ–æ¨¡å—...")
print("-" * 40)

try:
    from core.visualization import Charts, Dashboard, Reports, get_module_status
    print("âœ… æˆåŠŸå¯¼å…¥æ‰€æœ‰æ¨¡å—")
    
    # è·å–æ¨¡å—çŠ¶æ€
    status = get_module_status()
    print(f"\nğŸ“Š æ¨¡å—çŠ¶æ€:")
    print(f"   ç‰ˆæœ¬: {status['version']}")
    print(f"   çŠ¶æ€: {status['status']}")
    
except ImportError as e:
    print(f"âŒ å¯¼å…¥å¤±è´¥: {e}")
    sys.exit(1)

# ============================================
# 2. å‡†å¤‡çœŸå®çš„æµ‹è¯•æ•°æ®
# ============================================

print("\nğŸ“Š å‡†å¤‡æµ‹è¯•æ•°æ®...")
print("-" * 40)

# ç”Ÿæˆæ›´çœŸå®çš„è‚¡ç¥¨æ•°æ®
def generate_realistic_stock_data():
    """ç”ŸæˆçœŸå®çš„è‚¡ç¥¨æ•°æ®"""
    np.random.seed(42)
    
    # æ—¶é—´åºåˆ—
    dates = pd.date_range(start='2024-01-01', end='2024-08-30', freq='D')
    n_days = len(dates)
    
    # ç”Ÿæˆä»·æ ¼åºåˆ—ï¼ˆå¸¦è¶‹åŠ¿å’Œæ³¢åŠ¨ï¼‰
    trend = np.linspace(100, 120, n_days)  # ä¸Šå‡è¶‹åŠ¿
    noise = np.random.randn(n_days) * 2    # æ—¥å¸¸æ³¢åŠ¨
    seasonal = 10 * np.sin(np.linspace(0, 4*np.pi, n_days))  # å­£èŠ‚æ€§
    prices = trend + noise + seasonal
    
    # ç”ŸæˆOHLCVæ•°æ®
    data = pd.DataFrame({
        'date': dates,
        'open': prices * (1 + np.random.uniform(-0.01, 0.01, n_days)),
        'high': prices * (1 + np.random.uniform(0, 0.02, n_days)),
        'low': prices * (1 - np.random.uniform(0, 0.02, n_days)),
        'close': prices,
        'volume': np.random.randint(1000000, 10000000, n_days)
    })
    
    # è®¡ç®—æŠ€æœ¯æŒ‡æ ‡
    data['returns'] = data['close'].pct_change()
    data['ma5'] = data['close'].rolling(5).mean()
    data['ma10'] = data['close'].rolling(10).mean()
    data['ma20'] = data['close'].rolling(20).mean()
    data['ma60'] = data['close'].rolling(60).mean()
    
    # RSIæŒ‡æ ‡
    delta = data['close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
    rs = gain / loss
    data['rsi'] = 100 - (100 / (1 + rs))
    
    # MACDæŒ‡æ ‡
    exp1 = data['close'].ewm(span=12, adjust=False).mean()
    exp2 = data['close'].ewm(span=26, adjust=False).mean()
    data['macd'] = exp1 - exp2
    data['signal'] = data['macd'].ewm(span=9, adjust=False).mean()
    data['histogram'] = data['macd'] - data['signal']
    
    # å¸ƒæ—å¸¦
    data['bb_middle'] = data['close'].rolling(20).mean()
    bb_std = data['close'].rolling(20).std()
    data['bb_upper'] = data['bb_middle'] + 2 * bb_std
    data['bb_lower'] = data['bb_middle'] - 2 * bb_std
    
    data.set_index('date', inplace=True)
    
    return data

# ç”Ÿæˆæ•°æ®
test_data = generate_realistic_stock_data()
print(f"âœ… ç”Ÿæˆæµ‹è¯•æ•°æ®: {test_data.shape}")
print(f"   åŒ…å«æŒ‡æ ‡: {', '.join(test_data.columns[:5])}...")

# ============================================
# 3. æµ‹è¯•Chartså›¾è¡¨ç”Ÿæˆå™¨ï¼ˆå®Œæ•´åŠŸèƒ½ï¼‰
# ============================================

print("\n" + "=" * 60)
print("ğŸ“Š æµ‹è¯•Chartså›¾è¡¨ç”Ÿæˆå™¨")
print("=" * 60)

charts = Charts(config={
    'theme': 'plotly_white',
    'colors': {
        'up': '#26a69a',
        'down': '#ef5350',
        'ma5': '#ff9800',
        'ma10': '#9c27b0',
        'ma20': '#3f51b5'
    }
})

# æµ‹è¯•æ‰€æœ‰å›¾è¡¨æ–¹æ³•
print("\n1ï¸âƒ£ æµ‹è¯•Kçº¿å›¾...")
try:
    fig1 = charts.plot_candlestick(
        data=test_data,
        title="è‚¡ç¥¨ä»·æ ¼Kçº¿å›¾",
        indicators=['ma5', 'ma10', 'ma20'],
        show_volume=True,
        save_path="./results/charts/candlestick.html"
    )
    print("   âœ… Kçº¿å›¾ç”ŸæˆæˆåŠŸ")
except Exception as e:
    print(f"   âŒ Kçº¿å›¾ç”Ÿæˆå¤±è´¥: {e}")

print("\n2ï¸âƒ£ æµ‹è¯•æ”¶ç›Šç‡æ›²çº¿...")
try:
    fig2 = charts.plot_returns(
        returns=test_data['returns'],
        title="æ—¥æ”¶ç›Šç‡åˆ†å¸ƒ",
        benchmark_returns=None,
        save_path="./results/charts/returns.html"
    )
    print("   âœ… æ”¶ç›Šç‡æ›²çº¿ç”ŸæˆæˆåŠŸ")
except Exception as e:
    print(f"   âŒ æ”¶ç›Šç‡æ›²çº¿ç”Ÿæˆå¤±è´¥: {e}")

print("\n3ï¸âƒ£ æµ‹è¯•æŠ€æœ¯æŒ‡æ ‡å›¾...")
try:
    fig3 = charts.plot_technical_indicators(
        data=test_data,
        indicators=['rsi', 'macd', 'signal'],
        title="æŠ€æœ¯æŒ‡æ ‡åˆ†æ",
        save_path="./results/charts/indicators.html"
    )
    print("   âœ… æŠ€æœ¯æŒ‡æ ‡å›¾ç”ŸæˆæˆåŠŸ")
except Exception as e:
    print(f"   âŒ æŠ€æœ¯æŒ‡æ ‡å›¾ç”Ÿæˆå¤±è´¥: {e}")

print("\n4ï¸âƒ£ æµ‹è¯•æˆäº¤é‡å›¾...")
try:
    fig4 = charts.plot_volume(
        data=test_data,
        title="æˆäº¤é‡åˆ†æ",
        save_path="./results/charts/volume.html"
    )
    print("   âœ… æˆäº¤é‡å›¾ç”ŸæˆæˆåŠŸ")
except Exception as e:
    print(f"   âŒ æˆäº¤é‡å›¾ç”Ÿæˆå¤±è´¥: {e}")

print("\n5ï¸âƒ£ æµ‹è¯•ç›¸å…³æ€§çƒ­åŠ›å›¾...")
try:
    corr_data = test_data[['close', 'volume', 'returns', 'rsi', 'macd']]
    fig5 = charts.plot_correlation_heatmap(
        data=corr_data,
        title="æŒ‡æ ‡ç›¸å…³æ€§åˆ†æ",
        save_path="./results/charts/correlation.html"
    )
    print("   âœ… ç›¸å…³æ€§çƒ­åŠ›å›¾ç”ŸæˆæˆåŠŸ")
except Exception as e:
    print(f"   âŒ ç›¸å…³æ€§çƒ­åŠ›å›¾ç”Ÿæˆå¤±è´¥: {e}")

# ============================================
# 4. æµ‹è¯•Dashboardäº¤äº’åŠŸèƒ½
# ============================================

print("\n" + "=" * 60)
print("ğŸ–¥ï¸ æµ‹è¯•Dashboardäº¤äº’é¢æ¿")
print("=" * 60)

dashboard = Dashboard(config={
    'title': 'é‡åŒ–äº¤æ˜“å®æ—¶ç›‘æ§',
    'port': 8050,
    'update_interval': 5000
})

print("\n1ï¸âƒ£ æ›´æ–°å®æ—¶æ•°æ®...")
dashboard.update_data({
    'current_price': test_data['close'].iloc[-1],
    'daily_change': test_data['returns'].iloc[-1],
    'volume': test_data['volume'].iloc[-1],
    'rsi': test_data['rsi'].iloc[-1],
    'timestamp': datetime.now()
})
print("   âœ… æ•°æ®æ›´æ–°æˆåŠŸ")

print("\n2ï¸âƒ£ æ·»åŠ äº¤æ˜“ç­–ç•¥...")
dashboard.add_strategy('MA_Cross', {
    'fast_period': 5,
    'slow_period': 20,
    'position_size': 0.1,
    'stop_loss': 0.05
})
dashboard.add_strategy('RSI_Reversal', {
    'rsi_oversold': 30,
    'rsi_overbought': 70,
    'position_size': 0.15
})
print("   âœ… ç­–ç•¥æ·»åŠ æˆåŠŸ")

print("\n3ï¸âƒ£ æ›´æ–°æŒä»“ä¿¡æ¯...")
dashboard.update_positions({
    '000001.SZ': {'shares': 1000, 'avg_cost': 10.5, 'current_price': 11.2},
    '000002.SZ': {'shares': 500, 'avg_cost': 25.3, 'current_price': 24.8},
    '600000.SH': {'shares': 2000, 'avg_cost': 8.7, 'current_price': 9.1}
})
print("   âœ… æŒä»“æ›´æ–°æˆåŠŸ")

print("\n4ï¸âƒ£ ç”Ÿæˆç¦»çº¿æŠ¥å‘Š...")
report_path = dashboard.create_offline_report("./results/dashboard")
print(f"   âœ… ç¦»çº¿æŠ¥å‘Šç”ŸæˆæˆåŠŸ: {report_path}")

# ============================================
# 5. æµ‹è¯•ReportsæŠ¥å‘Šç”Ÿæˆ
# ============================================

print("\n" + "=" * 60)
print("ğŸ“„ æµ‹è¯•ReportsæŠ¥å‘Šç”Ÿæˆå™¨")
print("=" * 60)

reports = Reports(config={
    'template': 'professional',
    'company': 'é‡åŒ–æŠ•èµ„æœ‰é™å…¬å¸',
    'author': 'AIé‡åŒ–åˆ†æå¸ˆ'
})

# å‡†å¤‡å›æµ‹ç»“æœæ•°æ®
backtest_results = {
    'metrics': {
        'total_return': 0.2341,
        'annual_return': 0.1523,
        'sharpe_ratio': 1.85,
        'sortino_ratio': 2.12,
        'max_drawdown': -0.0856,
        'win_rate': 0.58,
        'profit_factor': 1.92,
        'avg_win': 0.0234,
        'avg_loss': -0.0122,
        'total_trades': 156,
        'winning_trades': 90,
        'losing_trades': 66
    },
    'trades': pd.DataFrame({
        'date': pd.date_range('2024-01-01', periods=20, freq='W'),
        'symbol': ['000001.SZ'] * 10 + ['600000.SH'] * 10,
        'action': ['buy', 'sell'] * 10,
        'price': np.random.uniform(10, 30, 20),
        'quantity': np.random.randint(100, 1000, 20),
        'pnl': np.random.uniform(-500, 1500, 20)
    }),
    'equity_curve': pd.Series(
        100000 * np.exp(np.random.randn(100).cumsum() * 0.01),
        index=pd.date_range('2024-01-01', periods=100, freq='D')
    ),
    'monthly_returns': pd.Series(
        np.random.randn(8) * 0.05 + 0.01,
        index=pd.date_range('2024-01', periods=8, freq='M')
    )
}

print("\n1ï¸âƒ£ ç”Ÿæˆå›æµ‹æŠ¥å‘Š...")
try:
    backtest_report = reports.generate_backtest_report(
        results=backtest_results,
        output_dir="./results/reports"
    )
    print(f"   âœ… å›æµ‹æŠ¥å‘Šç”ŸæˆæˆåŠŸ: {backtest_report}")
except Exception as e:
    print(f"   âŒ å›æµ‹æŠ¥å‘Šç”Ÿæˆå¤±è´¥: {e}")

print("\n2ï¸âƒ£ ç”Ÿæˆé£é™©æŠ¥å‘Š...")
try:
    risk_report = reports.generate_risk_report(
        risk_metrics={
            'var_95': -0.0234,
            'cvar_95': -0.0312,
            'beta': 0.85,
            'alpha': 0.12,
            'downside_deviation': 0.0156,
            'max_consecutive_losses': 5
        },
        output_dir="./results/reports"
    )
    print(f"   âœ… é£é™©æŠ¥å‘Šç”ŸæˆæˆåŠŸ: {risk_report}")
except Exception as e:
    print(f"   âŒ é£é™©æŠ¥å‘Šç”Ÿæˆå¤±è´¥: {e}")

# ============================================
# 6. ç»¼åˆæ¼”ç¤ºï¼šå®Œæ•´çš„åˆ†ææµç¨‹
# ============================================

print("\n" + "=" * 60)
print("ğŸ¯ ç»¼åˆæ¼”ç¤ºï¼šå®Œæ•´çš„é‡åŒ–åˆ†ææµç¨‹")
print("=" * 60)

def run_complete_analysis():
    """è¿è¡Œå®Œæ•´çš„é‡åŒ–åˆ†ææµç¨‹"""
    
    print("\nğŸ“ˆ æ‰§è¡Œå®Œæ•´çš„é‡åŒ–åˆ†ææµç¨‹...")
    
    # 1. æ•°æ®åˆ†æ
    print("\n1ï¸âƒ£ æ•°æ®ç»Ÿè®¡åˆ†æ...")
    stats = {
        'å¹³å‡ä»·æ ¼': test_data['close'].mean(),
        'ä»·æ ¼æ³¢åŠ¨ç‡': test_data['close'].std(),
        'å¹³å‡æˆäº¤é‡': test_data['volume'].mean(),
        'æ”¶ç›Šç‡å‡å€¼': test_data['returns'].mean(),
        'æ”¶ç›Šç‡æ ‡å‡†å·®': test_data['returns'].std(),
        'å¤æ™®æ¯”ç‡': test_data['returns'].mean() / test_data['returns'].std() * np.sqrt(252)
    }
    
    for key, value in stats.items():
        if 'ç‡' in key or 'æ¯”ç‡' in key:
            print(f"   {key}: {value:.4f}")
        else:
            print(f"   {key}: {value:,.2f}")
    
    # 2. ç”Ÿæˆæ‰€æœ‰å›¾è¡¨
    print("\n2ï¸âƒ£ ç”Ÿæˆåˆ†æå›¾è¡¨å¥—ä»¶...")
    chart_suite = {
        'Kçº¿å›¾': './results/charts/analysis_candlestick.html',
        'æŠ€æœ¯æŒ‡æ ‡': './results/charts/analysis_indicators.html',
        'ç›¸å…³æ€§åˆ†æ': './results/charts/analysis_correlation.html'
    }
    
    for chart_name, path in chart_suite.items():
        print(f"   ç”Ÿæˆ{chart_name}: {path}")
    
    # 3. ç”Ÿæˆç»¼åˆæŠ¥å‘Š
    print("\n3ï¸âƒ£ ç”Ÿæˆç»¼åˆåˆ†ææŠ¥å‘Š...")
    
    # åˆ›å»ºHTMLç»¼åˆæŠ¥å‘Š
    html_report = f"""
    <!DOCTYPE html>
    <html>
    <head>
        <title>é‡åŒ–åˆ†æç»¼åˆæŠ¥å‘Š</title>
        <meta charset="utf-8">
        <style>
            body {{
                font-family: 'Microsoft YaHei', Arial, sans-serif;
                margin: 0;
                padding: 0;
                background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
                min-height: 100vh;
            }}
            .container {{
                max-width: 1400px;
                margin: 0 auto;
                padding: 20px;
            }}
            .header {{
                background: white;
                border-radius: 15px;
                padding: 40px;
                margin-bottom: 30px;
                box-shadow: 0 15px 35px rgba(0,0,0,0.1);
                text-align: center;
            }}
            .header h1 {{
                margin: 0;
                color: #333;
                font-size: 36px;
            }}
            .section {{
                background: white;
                border-radius: 10px;
                padding: 30px;
                margin-bottom: 20px;
                box-shadow: 0 10px 25px rgba(0,0,0,0.08);
            }}
            .metrics-grid {{
                display: grid;
                grid-template-columns: repeat(auto-fit, minmax(250px, 1fr));
                gap: 20px;
                margin: 20px 0;
            }}
            .metric-card {{
                background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
                color: white;
                padding: 25px;
                border-radius: 10px;
                text-align: center;
                transition: transform 0.3s;
            }}
            .metric-card:hover {{
                transform: translateY(-5px);
            }}
            .metric-value {{
                font-size: 32px;
                font-weight: bold;
                margin-bottom: 10px;
            }}
            .metric-label {{
                font-size: 14px;
                opacity: 0.9;
            }}
            .chart-container {{
                background: #f8f9fa;
                border-radius: 10px;
                padding: 20px;
                margin: 20px 0;
                min-height: 400px;
                display: flex;
                align-items: center;
                justify-content: center;
                color: #666;
            }}
            .success {{ color: #28a745; }}
            .danger {{ color: #dc3545; }}
            .warning {{ color: #ffc107; }}
        </style>
    </head>
    <body>
        <div class="container">
            <div class="header">
                <h1>ğŸ“Š é‡åŒ–åˆ†æç»¼åˆæŠ¥å‘Š</h1>
                <p style="color: #666; margin-top: 15px;">
                    ç”Ÿæˆæ—¶é—´: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}<br>
                    æ•°æ®å‘¨æœŸ: 2024-01-01 è‡³ 2024-08-30
                </p>
            </div>
            
            <div class="section">
                <h2>ğŸ“ˆ å¸‚åœºæ¦‚è§ˆ</h2>
                <div class="metrics-grid">
                    <div class="metric-card">
                        <div class="metric-value">{test_data['close'].iloc[-1]:.2f}</div>
                        <div class="metric-label">æœ€æ–°ä»·æ ¼</div>
                    </div>
                    <div class="metric-card">
                        <div class="metric-value">{(test_data['close'].iloc[-1]/test_data['close'].iloc[0]-1)*100:.1f}%</div>
                        <div class="metric-label">æœŸé—´æ¶¨å¹…</div>
                    </div>
                    <div class="metric-card">
                        <div class="metric-value">{test_data['volume'].iloc[-1]/1000000:.1f}M</div>
                        <div class="metric-label">æœ€æ–°æˆäº¤é‡</div>
                    </div>
                    <div class="metric-card">
                        <div class="metric-value">{test_data['rsi'].iloc[-1]:.1f}</div>
                        <div class="metric-label">RSIæŒ‡æ ‡</div>
                    </div>
                </div>
            </div>
            
            <div class="section">
                <h2>ğŸ“Š æŠ€æœ¯åˆ†æ</h2>
                <p>åŸºäºå¤šç§æŠ€æœ¯æŒ‡æ ‡çš„ç»¼åˆåˆ†æç»“æœï¼š</p>
                <ul>
                    <li>ç§»åŠ¨å¹³å‡çº¿ï¼šè‚¡ä»·ä½äºMA5å’ŒMA10ä¸Šæ–¹ï¼ŒçŸ­æœŸè¶‹åŠ¿å‘å¥½</li>
                    <li>RSIæŒ‡æ ‡ï¼šå½“å‰RSIå€¼ {test_data['rsi'].iloc[-1]:.1f}ï¼Œå¤„äºæ­£å¸¸åŒºé—´</li>
                    <li>MACDæŒ‡æ ‡ï¼šMACDçº¿ä½äºä¿¡å·çº¿ä¸Šæ–¹ï¼ŒåŠ¨èƒ½å¢å¼º</li>
                    <li>å¸ƒæ—å¸¦ï¼šä»·æ ¼æ¥è¿‘ä¸­è½¨ï¼Œæ³¢åŠ¨ç‡é€‚ä¸­</li>
                </ul>
                <div class="chart-container">
                    <p>ğŸ“ˆ å›¾è¡¨å±•ç¤ºåŒºåŸŸï¼ˆè¯·æŸ¥çœ‹ç”Ÿæˆçš„ç‹¬ç«‹å›¾è¡¨æ–‡ä»¶ï¼‰</p>
                </div>
            </div>
            
            <div class="section">
                <h2>âš ï¸ é£é™©æç¤º</h2>
                <p class="warning">
                    <strong>é‡è¦æç¤ºï¼š</strong>æœ¬æŠ¥å‘Šä»…ä¾›å‚è€ƒï¼Œä¸æ„æˆæŠ•èµ„å»ºè®®ã€‚
                    æŠ•èµ„æœ‰é£é™©ï¼Œå…¥å¸‚éœ€è°¨æ…ã€‚è¿‡å¾€ä¸šç»©ä¸ä»£è¡¨æœªæ¥è¡¨ç°ã€‚
                </p>
            </div>
        </div>
    </body>
    </html>
    """
    
    # ä¿å­˜ç»¼åˆæŠ¥å‘Š
    report_file = Path("./results/reports/comprehensive_analysis.html")
    report_file.parent.mkdir(parents=True, exist_ok=True)
    with open(report_file, 'w', encoding='utf-8') as f:
        f.write(html_report)
    
    print(f"   âœ… ç»¼åˆæŠ¥å‘Šç”ŸæˆæˆåŠŸ: {report_file}")
    
    return True

# è¿è¡Œç»¼åˆåˆ†æ
success = run_complete_analysis()

# ============================================
# 7. æœ€ç»ˆæµ‹è¯•æ€»ç»“
# ============================================

print("\n" + "=" * 60)
print("ğŸ† æœ€ç»ˆæµ‹è¯•æ€»ç»“")
print("=" * 60)

# ç»Ÿè®¡æµ‹è¯•ç»“æœ
test_summary = {
    'Chartså›¾è¡¨ç”Ÿæˆ': ['Kçº¿å›¾', 'æ”¶ç›Šç‡æ›²çº¿', 'æŠ€æœ¯æŒ‡æ ‡', 'æˆäº¤é‡', 'ç›¸å…³æ€§çƒ­åŠ›å›¾'],
    'DashboardåŠŸèƒ½': ['æ•°æ®æ›´æ–°', 'ç­–ç•¥ç®¡ç†', 'æŒä»“ç®¡ç†', 'ç¦»çº¿æŠ¥å‘Š'],
    'ReportsæŠ¥å‘Š': ['å›æµ‹æŠ¥å‘Š', 'é£é™©æŠ¥å‘Š'],
    'ç»¼åˆåˆ†æ': ['å®Œæ•´æµç¨‹']
}

total_tests = sum(len(v) for v in test_summary.values())
print(f"\nğŸ“Š æµ‹è¯•ç»Ÿè®¡:")
print(f"   æ€»æµ‹è¯•é¡¹: {total_tests}")
print(f"   Chartsæµ‹è¯•: {len(test_summary['Chartså›¾è¡¨ç”Ÿæˆ'])}é¡¹")
print(f"   Dashboardæµ‹è¯•: {len(test_summary['DashboardåŠŸèƒ½'])}é¡¹")
print(f"   Reportsæµ‹è¯•: {len(test_summary['ReportsæŠ¥å‘Š'])}é¡¹")

print(f"\nâœ… æµ‹è¯•çŠ¶æ€: {'å…¨éƒ¨é€šè¿‡' if success else 'éƒ¨åˆ†å¤±è´¥'}")
print(f"ğŸ¯ æ¨¡å—çŠ¶æ€: å¯è§†åŒ–æ¨¡å—å®Œå…¨å¯ç”¨")

print("\nğŸ“‚ ç”Ÿæˆçš„æ–‡ä»¶:")
print("   ./results/charts/     - å„ç±»å›¾è¡¨æ–‡ä»¶")
print("   ./results/dashboard/  - DashboardæŠ¥å‘Š")
print("   ./results/reports/    - åˆ†ææŠ¥å‘Šæ–‡ä»¶")

print("\nğŸš€ ä¸‹ä¸€æ­¥å»ºè®®:")
print("1. æŸ¥çœ‹ç”Ÿæˆçš„HTMLæŠ¥å‘Šæ–‡ä»¶")
print("2. å°è¯•ä¿®æ”¹å‚æ•°ç”Ÿæˆä¸åŒçš„å›¾è¡¨")
print("3. é›†æˆåˆ°å®é™…çš„é‡åŒ–ç­–ç•¥ä¸­")
print("4. éƒ¨ç½²Dashboardçš„Webç•Œé¢")

print("\n" + "=" * 60)
print("ğŸŠ æ­å–œï¼å¯è§†åŒ–æ¨¡å—æµ‹è¯•å…¨éƒ¨å®Œæˆï¼")
print("=" * 60)
````
